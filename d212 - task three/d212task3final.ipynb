```python
# Remove empty rows
df = df[df['Presc01'].notna()]
```

```python
df.info()
```

```python
# dataframe converted to list of lists
```

```python
rows = []
for i in range(0, 7501):
    rows.append([str(df.values[i,j])
                 for j in range (0,20)])
```

```python
# list fed to TransactionEncoder

trans_encoder = TransactionEncoder()
fit_trans_encoder = trans_encoder.fit(rows).transform(rows)

# Return arrys to DataFrame
medications = pd.DataFrame(fit_trans_encoder, columns = trans_encoder.columns_)
```

```python
# looking
medications
```

```python
for col in medications.columns:
    print(col)
```

```python
# Remove empty column
cleaned_df = medications.drop(['nan'], axis = 1)
cleaned_df.head()
```

```python
# Write transformed data set to csv
cleaned_df.to_csv('d212task3clean.csv', index = False)
```

```python
df = cleaned_df
```

```python
df.shape
```

```python
# Creating Apriori object
rules = apriori(df, min_support = 0.02, use_colnames = True)
rules.head(5)
```

```python
# Creating rules table
rule_table = association_rules(rules, metric = 'lift', min_threshold = 1)
rule_table.head(20)
```

```python
# sorting rules by metric
```

```python
top_three_rules = rule_table.sort_values('lift', ascending=False).head(3)
top_three_rules
```
